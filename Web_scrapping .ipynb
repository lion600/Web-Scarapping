```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.99acres.com/sector-21-panchkula-overview-piffid"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
t = soup.title
title = t.text[:20]
title
```

```python
print(soup.find_all("p"))
```

```python
a = soup.find("p",class_="localityText__textContent localityText__hide").text
infra_index = a.find('Infrastructure')
rental_index = a.find('Rental Insights')
paragraph = a[:infra_index]
paragraph
```

```python

infra_heading = soup.find('p', class_='localityText__subHeading').text
infra_paragraph = soup.find_all('p', class_='localityText__textContent')[1].text
rental_heading = soup.find_all('p', class_='localityText__subHeading')[1].text
rental_div = soup.find('div', class_='localityText__hideForLandmarks')
rental_para= rental_div.find_all('p', class_='localityText__textContent')[1].text
```

```python
infra_paragraph
```

```python
data = {
    "Title":[title],
    "about":[paragraph],
    infra_heading: [infra_paragraph],
    rental_heading: [rental_para]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

# Scrapping News data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data = requests.get("https://inshorts.com/en/read")
data
```

```python
d = BeautifulSoup(data.content,"html.parser")
d
```

```python
headlines = d.find_all("span",attrs={'itemprop':'headline'})
headlines
```

```python
docs=[]
for i in headlines:
    docs.append(i.text)
    
docs    
```

```python
df1 = pd.DataFrame(docs,columns= ['NewsHeadlines'])
df1
```

```python
articles = d.find_all("div",attrs={'itemprop':'articleBody'})
articles
```

```python
docs1 = []
for a in articles:
    docs1.append(a.text)
    
docs1    
```

```python
df1['article']= docs1
```
